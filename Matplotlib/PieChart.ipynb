```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['month_name']=df['job_posted_date'].dt.month_name()
```

```python
df.groupby('job_title_short').size().plot(kind='pie',autopct='%1.1f%%',figsize=(10,10),startangle=90)
```

```python
new_df=df[['job_work_from_home','job_no_degree_mention','job_health_insurance']]
```

```python
fig , ax =plt.subplots(1,len(new_df.columns), figsize=(12, 4))
for i, col in enumerate(new_df.columns):
    new_df[col].value_counts().plot(kind='pie',ax=ax[i],autopct='%1.1f%%')
    ax[i].set_title(col)
    ax[i].set_ylabel("")
fig.tight_layout()
```
