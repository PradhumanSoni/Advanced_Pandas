```python

# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.head()
```

```python
final_data=df.groupby('job_title_short')['salary_year_avg'].median().sort_index()
final_data.plot(kind='barh')
plt.ylabel(" ")
plt.xlabel("Median Salary")
plt.title("Median Salary VS Jobs")
```
